```python
import pandas as pd

from sklearn.model_selection import train_test_split

df = pd.read_csv('Data/train.csv')
```

---

**Подготовка выборок: Sampling + Stratified Split 80/10/10**

```python
# Sampling: 1,5kk -> stratified 300k    
sample, _ = train_test_split(
    df,
    train_size=300000,
    stratify=df['category_name'],
    random_state=1337
)
```

```python
# Делим на 2 части: train 80%, test 20% 
train_df, temp_df = train_test_split(
    sample,
    test_size=0.2, 
    stratify=sample['category_name'],
    random_state=1337
)

# Также делим 20% на 2 части - val 10%, test 10%
val_df, test_df = train_test_split(
    temp_df,
    test_size=0.5,
    stratify=temp_df['category_name'],
    random_state=1337
)

print(len(train_df), len(val_df), len(test_df))
```

---

**Preprocessing: склеивание строк**

```python
for df in [train_df, val_df, test_df]:
    df['text'] = (df['title'].fillna('') + '. ' + df['description'].fillna('')).str.strip()

train_df['text'].head()
```

```python
train_df.to_parquet('parquets/avito_train.parquet', index=False)
val_df.to_parquet('parquets/avito_val.parquet', index=False)
test_df.to_parquet('parquets/avito_test.parquet', index=False)
```
